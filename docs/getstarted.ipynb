Guide d'utilisation pour l'API

Première étape ce rendre à l'adresse suivante : https://huggingface.co/spaces/leskimou/projet_MLops_part2

### 1. Authentification

Pour accéder à l'API vous devez avoir un identifiant et un mot de passe, si votre compte est bien enregistré dans la base de donnée alors vous pourrez vous connecter.

![authentification page](pics/auth.png)

### 2. La prédiction

Pour prédire si un client va réussir à rembourser son crédit, il suffit de renseigné l'ID du prêt. Qui correspond à la colonne SK_ID_CURR dans nos données. (Pour tester : cela va de 100002 à 456255)

![predit page](pics/preditsearch.png)

Une fois le bouton actionné il y a 3 possibilité :
* L'identifiant est invalide.

![id invalide](pics/preditidincorrect.png)


* Le score de la classe 1 (défaut de remboursement) dépasse le seuil, selon nos prédictions le prêt aura des problèmes de remboursement.

![predit defaut de remboursement](pics/preditdefault.png)


* Le score de la classe 1 (défaut de remboursement) ne dépasse pas le seuil, selon nos prédictions le prêt sera remboursé.

![predit remboursement](pics/preditok.png)

# 3. L'architecture

Pour le fonctionnement de cette API nous utilisons un fonctionnement par batch. Les prédictions pour chaque classe sont calculées en amont et envoyé à notre base de données (src\models\predict.py). L'API va permettre d'aller chercher les scors prédits pour chaque classe grâce à l'identifiant du prêt, puis tranche sur le résultat grâce au seuil optimisé lors de la création du modèle.

### L'administrateur

Si un compte avec le statut administrateur se connecte à l'API, il lui est possible d'avoir accès à l'onglet monitoring.

![onglet monitoring](pics/ongletmonito.png)

Cela donnes sur une interface de monitoring de l'API, l'objectif est de pouvoir suivre l'utilisation et les résultats renvoyé. On accède aussi au log de l'API avec différentes information sur les requetes qui ont été éffectué.

![monito page](pics/monitopage.png)
![monito logs](pics/monitologs.png)

# 4. Database

Notre solution repose sur une database hébergé en ligne par Supabse. Celle-ci contient 3 tables :
* Users, stock les informations relatif aux comptes pour avoir accès à l'API.
* Predictions, stock les scores de nos deux classes calculé par notre modèle.
* Predictions_logs, stock les informations relatif à la recherche d'une prédiction et l'utilisateur qui la réalise
